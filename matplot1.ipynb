## Area Plot

```python
from matplotlib import pyplot as plt
```

```python
days=[1,2,3,4,5,6]
reading=[12,21,13,24,15,26]
working=[2,4,6,8,10,12]
sleeping=[4,8,12,14,15,20]
dancing=[10,15,18,22,25,21]
```

```python
plt.plot([],[],color='r',label='Reading',linewidth=5)
plt.plot([],[],color='b',label='Sleeping',linewidth=5)
plt.plot([],[],color='y',label='Working',linewidth=5)
plt.plot([],[],color='g',label='Dancing',linewidth=5)
plt.stackplot(days,reading,working,sleeping,dancing,colors=['r','b','y','g'])
plt.title('Area Plot')
plt.xlabel('Student Infomation')
plt.ylabel('Roll')
plt.legend()
plt.show()
```
